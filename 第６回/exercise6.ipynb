# 属性の変更、削除（組み込み関数による方法）

6-1 で説明しましたが、属性の操作は「代入」だけではなく、以下のような組み込み関数を使うことができます。

---
```Python
setattr(オブジェクト, 属性文字列, 式)    # 追加・変更
delattr(オブジェクト, 属性文字列)        # 削除
```
---

これらの組み込み関数により、属性の名前がプログラムを実行してから定まるときにも、属性の操作することができます。
すなわち、上の属性文字列を書くべきところで、文字列に評価される式を書くことができます。

## 3. `UTCourse` オブジェクトの利用

````python
import utas
uc = utas.UTCourse({"year":2018, "title_j":"コンピュータシステム概論", "name_j":"小林克志", "Teaching_Methods": "講義では...."})

print(type(uc.year))
print(uc.title_j)
````

````python
import json
import utas
def utcourses():
    uta=[]
    with open("catalog-2018.json","r") as f:
        j=json.load(f)
        for jj in j:
            ut=utas.UTCourse(jj)
            uta.append(ut)
    return uta
utcs = utcourses()

#print(type(utcs[0].year))
print(utcs)
````

````python
import utas
def filter_ccc(s, utcs):
    classes=[]
    for utc in utcs:
        if hasattr(utc,"Common_Course_Code") and utc.Common_Course_Code==s:
            classes.append(utc)
    return classes


utcs = utcourses()                      # json ファイルの読み込み
data = filter_ccc("FEN-MP4d01L1", utcs)
print(data)
````
